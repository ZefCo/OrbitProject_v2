Principle of Lazyness

```python
import pandas
import pathlib
import re
import numpy as np
# import plotly.graph_objects as go
import datetime
from matplotlib import pyplot as plt

cwd = pathlib.Path.cwd()
```

Imports the data

```python
orbit_table = pandas.read_csv(cwd / "Orbit_Table.csv", header = 0)
orbit_table = orbit_table.set_index("t")
# print(orbit_table)
```

Plots the graphs

```python
rows, cols = orbit_table.shape

for c in range(int(cols / 2)):
    cndex = (2*c, 2*c+1)
    col_of_interest: pandas.DataFrame = orbit_table.iloc[:, cndex[0]:cndex[1] + 1].copy()
    # print(col_of_interest.iloc[:, 0])
    name = re.split("_", tuple(col_of_interest.columns)[0])[0]

    plt.plot(col_of_interest.iloc[:, 0], col_of_interest.iloc[:, 1])

plt.show()
```
