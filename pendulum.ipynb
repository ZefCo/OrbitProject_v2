```python
import numpy as np
import matplotlib.pyplot as plt


def f(r, t):
    th, w = r
    fth = w
    fw = -np.sin(th)
    return np.array([fth, fw])

def RK4(f, x, t, h):
    k1 = h*f(x, t)
    k2 = h*f(x + k1/2, t + h/2)
    k3 = h*f(x + k2/2, t + h/2)
    k4 = h*f(x + k3, t + h)
    return x + (k1 + 2*k2 + 2*k3 + k4)/6
```

```python
delta = 1e-5
tmax = 20
h = 0.001

t_values = []
th_values = []
w_values = []

t = 0
th, w = 3.1, 0

while t < tmax:
    t_values.append(t)
    th_values.append(th)
    w_values.append(w)
    rho = 0
    while rho < 1:
        dt = 2*h
        th1, w1 = RK4(f, (th, w), t, h)
        th1, w1 = RK4(f, (th1, w1), t+h, h)
        th2, w2 = RK4(f, (th, w), t, 2*h)
        eps = abs(th1 - th2)/30
        rho = h*delta/eps
        h = min(h*rho**(1/4), 2*h)
    th, w = th1, w1
    t += dt
    # print(t)

plt.plot(t_values, th_values, "o")
plt.show()
```
